```python
data.to_excel('Dataset_clean.xlsx')
data_clean = pd.read_excel('Dataset_clean.xlsx')
data_clean
```

```python
data_clean = data_clean.astype({'Kelayakan' : 'category'})
data_clean = data_clean.astype({'Review' : 'string'})
data_clean.dtypes
```

```python
#TF-IDF
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
text_tf = tf.fit_transform(data_clean['Review'].astype('U'))

for i in text_tf:
  print(i)
```

```python
#splitting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(text_tf, data_clean['Kelayakan'], test_size = 0.2, random_state = 0)
```

```python
#naive bayes

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

clf = MultinomialNB().fit(X_train, y_train)
predicted = clf.predict(X_test)

print("MultinomialNB Accuracy:", accuracy_score(y_test, predicted))
print("MultinomialNB Precision:", precision_score(y_test, predicted, average = 'binary', pos_label='Tidak'))
print("MultinomialNB Recall:", recall_score(y_test, predicted, average = 'binary', pos_label='Tidak'))
print("MultinomialNB f1_score:", f1_score(y_test, predicted, average = 'binary', pos_label='Tidak' ))

print(f'confusion_matrix:\n {confusion_matrix(y_test, predicted)}')
print("========================================================================================")
print(classification_report(y_test, predicted))
```
